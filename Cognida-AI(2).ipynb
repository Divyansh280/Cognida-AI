```python
#2. Finding the area of all the irregular images
```

```python
#Using basic approach of brute force method and calculating all the pixels to find the area of the figure
area = 0
for i in range(len(input_image1)):
    for j in range(len(input_image1[0])):
        for k in range(len(input_image1[0][0])):
            if input_image1[i][j][k]==0:
                area+=1
#Displaying the image with answer
str1 = "Total area of the image:" + str(area)
plt.imshow(input_image1)
plt.text(350,200,str1, color = "red")
```
