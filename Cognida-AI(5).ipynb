```python
#Import the required libraries
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
#Reading the input files using CV2 library
input_image1 = cv2.imread(r"C:\Users\ayush\OneDrive\Desktop\MVSR_2451-20-733-058_Divyansh_Jain_CCC_MinTech\input\img1.png")
input_image2 = cv2.imread(r"C:\Users\ayush\OneDrive\Desktop\MVSR_2451-20-733-058_Divyansh_Jain_CCC_MinTech\input\img2.png")
input_image3 = cv2.imread(r"C:\Users\ayush\OneDrive\Desktop\MVSR_2451-20-733-058_Divyansh_Jain_CCC_MinTech\input\img3.png")
input_image4 = cv2.imread(r"C:\Users\ayush\OneDrive\Desktop\MVSR_2451-20-733-058_Divyansh_Jain_CCC_MinTech\input\img4.png")
input_image5 = cv2.imread(r"C:\Users\ayush\OneDrive\Desktop\MVSR_2451-20-733-058_Divyansh_Jain_CCC_MinTech\input\img5.png")
```

```python
#5. Finding the centroid of the image
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(input_image1, cv2.COLOR_BGR2GRAY)

ret,thresh = cv2.threshold(gray,150,255,0)

# Find the contours using binary image
contours1,hierarchy = cv2.findContours(thresh, cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
cnt = contours1[1]


# Calculate the moments of the largest contour
M = cv2.moments(cnt)

# Calculate the centroid coordinates
centroid_x = int(M['m10'] / M['m00'])
centroid_y = int(M['m01'] / M['m00'])

# Draw a circle at the centroid position
cv2.circle(input_image1, (centroid_x, centroid_y), 5, (0, 0, 255), -1)  # (0, 0, 255) represents red color

str2 = "centroid" + "(" + str(centroid_x) + "," + str(centroid_y) + ")"

# Output the centroid coordinates
plt.imshow(input_image1)
plt.text(200,20,str2, color = "red")
```
