```python
def find_smallest_enclosing_circle(image):
  # Convert the image to grayscale.
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  # Apply a threshold to the image to binarize it.
    thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

  # Find the contours in the binary image.
    contours = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]

  # For each contour, find the minimum bounding rectangle.
    bounding_rectangles = []
    for contour in contours:
        bounding_rectangles.append(cv2.minAreaRect(contour))

  # For each bounding rectangle, calculate the radius of the smallest circle that
  # encloses the rectangle.
    smallest_circle = None
    for bounding_rectangle in bounding_rectangles:
        width, height, angle = bounding_rectangle[0][1], bounding_rectangle[1][1], bounding_rectangle[2]
        radius = np.sqrt(width**2 + height**2) / 2
    if smallest_circle is None or radius < smallest_circle[2]:
        smallest_circle = (bounding_rectangle[0][0], bounding_rectangle[0][1], radius)

    return smallest_circle


# Find the smallest circle that encloses the particle in the image.
smallest_circle = find_smallest_enclosing_circle(input_image1)

# Convert the radius to an integer.
radius_int = int(smallest_circle[2])

# Draw the circle on the image.
cv2.circle(input_image1, (int(smallest_circle[0]), int(smallest_circle[1])), radius_int, (0, 0, 255), 2)

# Display the image.
plt.imshow(input_image1)
```
