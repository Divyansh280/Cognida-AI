```python
#4. Finding the perimeter of the irregular figure
```

```python
gray = cv2.cvtColor(input_image1, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (3, 3), 0)

edged = cv2.Canny(blurred, 10, 100)

# Apply thresholding in the gray image to create a binary image
ret,thresh = cv2.threshold(gray,150,255,0)

# Find the contours using binary image
contours1,hierarchy = cv2.findContours(thresh, cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
cnt = contours1[1]

# compute the perimeter
perimeter = cv2.arcLength(cnt, True)
perimeter = round(perimeter, 4)

plt.imshow(edged)
plt.text(350,200,perimeter, color = "red")
```
